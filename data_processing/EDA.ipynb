```python
from google.colab import drive
import os
import matplotlib.pyplot as plt

# mount Google Drive
drive.mount('/content/gdrive')
```

```python
directory_path = '/content/gdrive/My Drive/APS360 Project/Output'

class_names = []
image_counts = []

# loop through subdirectories and count images
for subdir in os.listdir(directory_path):
    subdir_path = os.path.join(directory_path, subdir)
    if os.path.isdir(subdir_path):
        image_count = sum(1 for f in os.listdir(subdir_path) if f.endswith('.jpg'))
        class_names.append(subdir)
        image_counts.append(image_count)


plt.bar(class_names, image_counts)
plt.title("Image Counts by Class - Before Filtering")
plt.xlabel("Class (City)")
plt.ylabel("Image Count")
plt.xticks(rotation = 90)
plt.show()
```
